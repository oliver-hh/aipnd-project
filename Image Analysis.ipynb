````python
# Imports used for the analysis
import glob
import json
import pandas as pd
import numpy as np
from PIL import Image
````

## Image Statistics

````python
files = glob.glob('./flowers/**/*.jpg', recursive=True)  
sizes = [Image.open(file).size for file in files]  
  
widths, heights = zip(*sizes)  
  
n = len(files)  
print(f"Number of images: {n}")  
print(f"Avg width = {int(np.mean(widths))}")  
print(f"Avg height = {int(np.mean(heights))}")  
print(f"Min width = {min(widths)}")  
print(f"Min height = {min(heights)}")  
print(f"Max width = {max(widths)}")  
print(f"Max height = {max(heights)}")  
````

## Show image categories sorted in a Data Frame

````python
with open('./cat_to_name.json', 'r') as f:
    categories_json = json.load(f)

categories_array = [{'id': int(k), 'name': v} for k, v in categories_json.items()]
categories_array.sort(key=lambda x: x['id'])
#print(categories_array)

categories_df = pd.DataFrame(categories_array).set_index('id')
pd.set_option('display.max_rows', None)
#pd.reset_option('display.max_rows')

print(f"Number of Categories: {len(categories_df)}")
categories_df
````

## Reorganize categories file
For clarity the file ```cat_to_name.json``` should be saved ordered by ```id```.

````python
sorted_data = dict(sorted(categories_json.items(), key=lambda x: int(x[0])))  
with open('./cat_to_name.json', 'w') as f:  
    json.dump(sorted_data, f, indent=4) 
````
